# Análise dos dados de escolas Municipais em situação precária, utilizando API

São dados do INEP

Análise dos dados obtidos pela API da Educação Inteligente



Origem: http://educacao.dadosabertosbr.com/api/docs/%2Fapi%2Fescolas%2Fbuscaavancada

URL Request: http://educacao.dadosabertosbr.com/api/escolas/buscaavancada?situacaoFuncionamento=1&energiaInexistente=on&aguaInexistente=on&esgotoInexistente=on&cozinha=on


Utilizando API, as atualizações ocorreram conforme o site atualizar;

Os dados estão sendo passado por JSON;


Tem 0 a 39 linhas na data 14/07/2020

```python
#Importando as bibliotecas

import pandas as pd
import requests as rq
import numpy as np

url =  'http://educacao.dadosabertosbr.com/api/escolas/buscaavancada?situacaoFuncionamento=1&energiaInexistente=on&aguaInexistente=on&esgotoInexistente=on&cozinha=on'

resposta = rq.get(url)

resposta.text
```

```python
 '''Obtendo os dados do dicionário, observe que não tem chave e valor.
    Nesse caso pode-se usar o valor 0 que retorna 538
    ou o valor 1 que retorna todos os dados.
    
    Está acessando o dicionário por indice
    
 '''


df = pd.DataFrame(resposta.json()[1])
df.head(20)
```
